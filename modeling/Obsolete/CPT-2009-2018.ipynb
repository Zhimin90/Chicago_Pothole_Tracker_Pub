import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st


```python
grid_size = 100
from scipy import stats as st

def get_kde(geo_price_map_filtered, xmin, xmax, ymin, ymax):
    # Extract x and y
    #print(geo_price_map_filtered)
    x = geo_price_map_filtered["LONGITUDE"].dropna().to_numpy()
    y = geo_price_map_filtered["LATITUDE"].dropna().to_numpy()
    xx, yy = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
    positions = np.vstack([xx.ravel(), yy.ravel()])
    values = np.vstack([x, y])
    print("values is: " + str(values))
    kernel = st.gaussian_kde(dataset=values, bw_method="silverman" )
    kernel.set_bandwidth(bw_method=kernel.factor / 30.)
    f = np.reshape(kernel(positions).T, xx.shape)
    print(f)
    return f
```

```python
density_matrix_t_series = []
# Define the borders
x = map_arr[-1]["LONGITUDE"].dropna().to_numpy()
y = map_arr[-1]["LATITUDE"].dropna().to_numpy()
deltaX = (max(x) - min(x))/10
deltaY = (max(y) - min(y))/10
xmin = min(x) - deltaX
xmax = max(x) + deltaX
ymin = min(y) - deltaY
ymax = max(y) + deltaY
xx, yy = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
for i, df in enumerate(map_arr):
    if df["LONGITUDE"].count() > 10:
        density_matrix_t_series.append(get_kde(df, xmin, xmax, ymin, ymax))
        print("@" + str(i))
    
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM
```

```python
from sklearn.preprocessing import MinMaxScaler
np.seterr(divide='ignore', invalid='ignore')
scaler = MinMaxScaler()
normalized_matrices = []
#normalized_matrices = np.array(normalized_matrices, dtype=np.float64)
for matrix in density_matrix_t_series:
    normalized_matrices.append(scaler.fit_transform(matrix))
```

```python
np.max(normalized_matrices[14])
```

```python
for series in density_matrix_t_series:
    print(np.max(series))
```

```python
for i in range(len(normalized_matrices)):
    normalized_matrices[i] = density_matrix_t_series[i] / 4558.8114765635955
```

```python
new_matric = []
for i in range(len(normalized_matrices)):
    new_matric.append(normalized_matrices[i].flatten())
x_train = np.array(normalized_matrices[0:-2].copy())
y_train = np.array(new_matric[1:-1].copy())
```

```python
model = Sequential()
model.add(LSTM(100, input_shape=x_train.shape[1:], activation='tanh', return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100, activation='tanh'))
model.add(Dropout(0.1))
model.add(Dense(100, activation='tanh'))
model.add(Dropout(0.1))

model.add(Dense(10000, activation='softmax'))
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(lr=0.001, decay=1e-6)

model.compile(
    loss='mse',
    optimizer=opt,
    metrics=['mae'],
)

model.fit(x_train,
          y_train,
          epochs=50)
```
